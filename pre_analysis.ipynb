```python
df_features = df2[['ppagecat', 'ppeducat', 'ppgender', 'ppincimp', 'qflag', 'glbstatus', 'married', 'q24_met_online']] 
df_features.shape
```

```python
df_cleaned = df_features.dropna()
df_cleaned.shape
```

```python
from sklearn.cluster import KMeans
import numpy as np

kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
labels = kmeans.fit_predict(df_cleaned)

print("Clusters", labels)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import numpy as np

kmeans = KMeans(n_clusters=3, random_state=42).fit(df_cleaned)
labels = kmeans.labels_

plt.scatter(df_cleaned[:, 0], df_cleaned[:, 1], c=labels, cmap='viridis')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("K-Means Clustering")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Видалення нечислових стовпців (або їх кодування)
df_cleaned = df_cleaned.select_dtypes(include=['number']).dropna()

# Масштабування (нормалізація)
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_cleaned), columns=df_cleaned.columns)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Визначення оптимальної кількості кластерів за методом "Elbow"
inertia = []
K_range = range(2, 10)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Графік для визначення оптимального k
plt.plot(K_range, inertia, marker='o')
plt.xlabel('Кількість кластерів (k)')
plt.ylabel('Інерція')
plt.title('Метод ліктя (Elbow Method)')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df_cleaned['Cluster'] = kmeans.fit_predict(df_scaled)
```

```python
from sklearn.decomposition import PCA
import seaborn as sns

pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)

df_visual = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])
df_visual['Cluster'] = df_cleaned['Cluster']

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_visual, x='PC1', y='PC2', hue='Cluster', palette='viridis')
plt.title("Кластеризація HCMST (PCA)")
plt.show()
```
